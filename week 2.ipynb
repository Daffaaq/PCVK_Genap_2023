Masukkan Library dan dilanjutkan dengan membuka file Google Drive yang 
sudah ada. Contohnya pada kode program di bawah file image dengan nama

````python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/PVCK/Pertemuan 2/Kitten.jpg')
plt.imshow(img)                          #perhatikan hasilnya adalah citra dgn channerl warna RGB
````

OpenCV membaca image dan menyimpan dalam channel warna BGR (Blue Green Red) 
Tampilkan nilai pixel dan image 


````python
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB) #konversi channel BGR -> RGB
plt.imshow(img3)
````

Menampilkan citra Grayscale, melakukan resizing, melakukan Flipping, Menyimpan 
citra hasil

````python
img_gray = cv.imread('/content/drive/MyDrive/PVCK/Pertemuan 2/Kitten.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray) 
````

Dari citra grayscale, selanjutnya dipilih untuk ditampilkan colormap dengan warna 
‘gray’ 

````python
plt.imshow(img_gray, cmap='gray')
````

Dari citra grayscale, selanjutnya dipilih untuk ditampilkan colormap dengan warna 
‘magma’

````python
plt.imshow(img_gray, cmap='magma')
````

Citra RGB di resize dalam ukuran panjang 512 dan lebar 1024 

````python
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512,1024))
plt.imshow(img4)
````

Citra RGB ditampilkan dalam ukuran yang lebih besar dengan posisi gambar terbalik 

````python
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB),0)

#simpan file image
#cv.imwrite('/content/drive/MyDrive/PCVK_MM/')

#tampilkan plot dengan ukuran canvas yang lebih besar
fig = plt.figure(figsize=(10,10))  #ubah ukuran (10,10) sesuai kebutuhan
ax = fig.add_subplot(111)
ax. imshow(img5) 
````

Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image 
dengan tipe data int16.

````python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
````

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2 

````python
#perhatikan koordinat titik2 ptl dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
````

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 
yang tertulis pada kode program. 

````python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
````

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang 
tertulis pada kode program.

````python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
````

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut. 

````python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
````

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah 
ditentukan.

````python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,
           color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img) 
````

Pembuatan black image kembali dilakukan dengan tipe data int32 

````python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
````

Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32 

````python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
````

Array tersebut kemudian di reshape sebagai berikut 

````python
pts = vertices.reshape((-1,1,2)) #nilai untuk menunjukan bahwa tiap titik dibuat 3 channel yang memiliki R, G dan B
pts
````

Penambahan polyline pada black image kedua yang telah dibuat. 

````python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
````
